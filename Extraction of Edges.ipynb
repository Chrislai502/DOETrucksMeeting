```python
# This function will store the data given from file "edge_cost.csv" into multiple csv files.
# To be accurate, the number of files generated will be:
# (Total CSV Files) = (# of vehiecles) * (# of Vehiecle Types) * (# of days) * (# of states) 

# Question: 
# Handling edge cases where the edge have only 1 data
# Is the variance of only 2 datapoints actually reliable
# (a -> b), (b -> a) not grouped together

def generateData (folderName, days = 1, numVehiecles = 1, randomSeed = 1):
    variance = 0.15
    np.random.seed(randomSeed)
    df = pd.read_csv('edge_cost.csv')
    parent_dir = os.getcwd()
    savePath = os.path.join(parent_dir, folderName)
    
    # Getting information from the table
    vehiecleTypes = df.Type.unique()
    numNodes = max(df.Start_Node.max(), df.End_Node.max())
#     sList = ["s1", "s2", "s3", "s4", "s5", "s6"]
    cols = ["i_" + str(i) for i in range(1, numNodes+1)]
    index = ["j_" + str(i) for i in range(1, numNodes+1)]
                   
    # Get the edge mean for vehiecle
    for v in vehiecleTypes:
        tempPath = os.path.join(savePath, v)
        workFrame = df[df["Type"] == v]
        test_mean = []
        # Iterate over Days
        for d in range(1, days + 1) :
            tempPath1 = os.path.join(tempPath, "d" + str(d))

            # Delete the contents with shutil if it already exists
            if os.path.exists(tempPath1):
                shutil.rmtree(tempPath1)
            # Create the path with OS
            os.makedirs(tempPath1)
            # In final table, goal is to have State vs (NumNodes C 2)# of
            # tuples (mean, std) as csv 
#             cols = list(combinations(range(1, numNodes + 1), 2))
            newFrame = pd.DataFrame(columns = cols, index = index)
                   
            for s in range(1, 7):
                state = "State" + str(s) + "_kwh_mile"
#                 tempPath2 = os.path.join(tempPath1, "state" + str(s))
                
                # Iterate over node values.
                for i in range(1, numNodes + 1):
                    for j in range(1, numNodes + 1):
                        # This line below gives a frame of the cost of an edge
                        print("\nFor node (%d, %d):\nData:" % (i, j) )
                        data = workFrame.loc[((workFrame['Start_Node'] == i) & (workFrame['End_Node'] == j))][state]
                        
                        # edge case if i = j
                        if data.size == 0:
                            newFrame.iat[i-1, j-1] = -1
                            continue
                        print("Data Size: ", data.size)
                        mean = data.values[0]
                        stdv = variance * mean
                        print("Mean: ", mean, ", stdv: ", stdv)

                        # convention of smaller value first
                        sample = np.random.lognormal(mean, stdv)
                        sampleScipy = stats.lognorm(s = stdv, scale = mean)
                        print("Sample Numpy: ", sample)
                        print("Sample Scipy: ", sampleScipy.rvs()) #rvs stands for random value sampling
                        newFrame.iat[i - 1, j - 1] = sampleScipy.rvs()

                display(newFrame)
                

                # Create the filePath
                tempPath2 = os.path.join(tempPath1, "state" + str(s) + ".csv")
                
#                 tempPath2 = os.path.join(tempPath2, "state" + str(s) + ".csv")
                print("PAth: ", tempPath2)
                newFrame.to_csv(tempPath2)
    print(test_mean)
```

```python
# Reads out the data and collects them into an array
def readDataCSV (folderName = ""):
    # Initialize some information
    numDays = 0
    path = os.path.join(os.getcwd(), folderName)
    VehiecleTypes = os.listdir(path)
    numNodes = 0
    arr = {}

    for v in VehiecleTypes:
        tempPath1 = os.path.join(path, v)
        days = os.listdir(tempPath1)
        arr[v] = [] # This needs to change, initialize "Days number of empty lists"
        for d, filename in enumerate(days):
            arr[v].append([])
            tempPath2 = os.path.join(tempPath1, filename)
            tempPath2 = os.path.join(tempPath2, filename + ".csv")
            df = pd.read_csv(tempPath2)
            df = df.iloc[: , 1:]

            # helper function to find the numbers in a string
            def findInt(sentence):
                return [int(i) for i in sentence if i.isdigit()]
#             print(df.columns)
            headerVals = findInt("".join(df.columns))
#             print(headerVals)
            numNodes = max(headerVals)
            from_to = np.full((numNodes, numNodes), 0)
            
            # Iterating over all the nodes
            for column in df:
                print(df[column].values)
                ij = findInt(column)
                print(ij)
                i = ij[0] - 1
                j = ij[1] - 1
                ij_col_vals = df[column].values
                
                # Put s for later because s is already fixed,so iterating it after the columns might be easier
                for s in range(6):
                    val = ij_col_vals[s]
                    
                    # how to deal with from node to node
                    from_to[i][j] = val
                    from_to[j][i] = val
            arr[v][d].append(from_to) # where is the states dimension?

    return arr
```
